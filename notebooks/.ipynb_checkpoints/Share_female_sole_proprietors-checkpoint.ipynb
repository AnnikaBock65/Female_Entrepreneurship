```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
female_sole_pro = pd.read_csv('../data/Share_of_female_sole_proprietors.csv', engine='python')
female_sole_pro.shape
```

```python
female_sole_pro
```

```python
female_sole_pro.info()
```

```python
female_sole_pro.isnull().sum()
```

```python
female_sole_pro_drop = female_sole_pro.dropna()
```

```python
female_sole_pro_drop
```

```python
female_sole_pro_renamed = female_sole_pro_drop.rename(columns = {'Income group':'Income_group', 'Share of female sole proprietors':'Share_of_female_sole_proprietors' })
```

```python
female_sole_pro_renamed
```

```python
female_sole_pro_renamed.info()
```

```python
female_sole_pro_renamed['Share_of_female_sole_proprietors'] = female_sole_pro_renamed['Share_of_female_sole_proprietors'].str.replace('%', '')
```

```python
female_sole_pro_renamed
```

```python
female_sole_pro_renamed.info()
```

```python
female_sole_pro_renamed['Share_of_female_sole_proprietors'] = female_sole_pro_renamed['Share_of_female_sole_proprietors'].astype(float)
```

```python
female_sole_pro_renamed.info()
```

```python
female_sole_pro_renamed
```

```python
female_sole_pro_renamed.Income_group.value_counts()
```

```python
High_income = female_sole_pro_renamed.query('Income_group == "High income"')
High_income.head()
```

```python
Low_income = female_sole_pro_renamed.query('Income_group == "Low income"')
Low_income.head()
```

```python
Lower_middle_income = female_sole_pro_renamed.query('Income_group == "Lower middle income"')
Lower_middle_income.head()
```

```python
Upper_middle_income = female_sole_pro_renamed.query('Income_group == "Upper middle income"')
Upper_middle_income.head()
```

```python
High_income_average = (High_income.groupby(['Income_group'], as_index=False)['Share_of_female_sole_proprietors'].mean())
High_income_average
```

```python
Low_income_average = (Low_income.groupby(['Income_group'], as_index=False)['Share_of_female_sole_proprietors'].mean())
Low_income_average
```

```python
Lower_middle_income_average = (Lower_middle_income.groupby(['Income_group'], as_index=False)['Share_of_female_sole_proprietors'].mean())
Lower_middle_income_average
```

```python
Upper_middle_income_average = (Upper_middle_income.groupby(['Income_group'], as_index=False)['Share_of_female_sole_proprietors'].mean())
Upper_middle_income_average
```

```python
High_income['difference_to_average_share'] = High_income['Share_of_female_sole_proprietors'] - High_income_average['Share_of_female_sole_proprietors'].values[0]
```

```python
High_income.head()
```

```python
Low_income['difference_to_average_share'] = Low_income['Share_of_female_sole_proprietors'] - Low_income_average['Share_of_female_sole_proprietors'].values[0]
```

```python
Low_income.head()
```

```python
Lower_middle_income['difference_to_average_share'] = Lower_middle_income['Share_of_female_sole_proprietors'] - Lower_middle_income_average['Share_of_female_sole_proprietors'].values[0]
```

```python
Lower_middle_income.head()
```

```python
Upper_middle_income['difference_to_average_share'] = Upper_middle_income['Share_of_female_sole_proprietors'] - Upper_middle_income_average['Share_of_female_sole_proprietors'].values[0]
```

```python
Upper_middle_income.head()
```

```python
Upper_middle_income.info()
```

```python
concat_data = pd.concat([High_income, Low_income, Lower_middle_income, Upper_middle_income])
concat_data
```

```python
concat_data.to_csv('diff_share_female_sole_pro.csv')
```

```python
concat_data.Year.value_counts()
```

```python
df_all_14_18 = concat_data.query('Year == 2014 or Year == 2018')
df_all_14_18.head()
```

```python
merge_18_14
```

```python
merge_18_14.info()
```

```python
merge_18_14['development_diff_share'] = merge_18_14['difference_to_average_share_y'] - merge_18_14['difference_to_average_share_x']
```

```python
merge_18_14.head()
```

```python
merge_18_14_drop = merge_18_14.dropna()
```

```python
merge_18_14_drop
```

```python
merge_18_14_drop.info()
```

```python
merge_18_14_drop['Year_x'] = merge_18_14_drop['Year_x'].astype('int')
```

```python
merge_18_14_drop.info()
```

```python
merge_18_14_drop['Year_y'] = merge_18_14_drop['Year_y'].astype('int')
```

```python
merge_18_14_drop.info()
```

```python
merge_18_14_drop
```

```python
merge_18_14_drop = merge_18_14_drop.rename(columns = {'Year_x':'Year_2018', 'Year_y':'Year_2017', 'difference_to_average_share_x':'difference_to_average_share_2018', 'difference_to_average_share_y':'difference_to_average_share_14'  })
```

```python
merge_18_14_drop
```

```python
merge_18_14_drop = merge_18_14_drop.rename(columns = {'Share_of_female_sole_proprietors_x':'Share_of_female_sole_proprietors_2018', 'Year_y':'Year_2017', 'difference_to_average_share_x':'difference_to_average_share_2018', 'Share_of_female_sole_proprietors_y':'Share_of_female_sole_proprietors_2014'  })
```

```python
merge_18_14_drop
```

```python
df_development_14_18 = merge_18_14_drop.sort_values(by=['development_diff_share'], ascending = False)
```
