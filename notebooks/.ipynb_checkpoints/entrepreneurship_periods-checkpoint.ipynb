```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df_entre_periods = pd.read_csv('../data/entrepreneurship_periods.csv', engine='python')
df_entre_periods.shape
```

```python
df_entre_periods.head()
```

```python
def law_count(counter):
    if (counter['sign_contract'] == "Yes") | (counter['register_business'] == "Yes") | (counter['open_bank_account'] == "Yes") | (counter['discrimination_credit_prohibited'] == "Yes") :
        return '4'
    if (counter['sign_contract'] == "Yes") | (counter['register_business'] == "No") | (counter['open_bank_account'] == "Yes") | (counter['discrimination_credit_prohibited'] == "Yes") :
        return '3'
    if (counter['sign_contract'] == "Yes") | (counter['register_business'] == "Yes") | (counter['open_bank_account'] == "No") | (counter['discrimination_credit_prohibited'] == "Yes") :
        return '3'
    if (counter['sign_contract'] == "Yes") | (counter['register_business'] == "Yes") | (counter['open_bank_account'] == "Yes") | (counter['discrimination_credit_prohibited'] == "No") :
        return '3'
    if (counter['sign_contract'] == "No") | (counter['register_business'] == "No") | (counter['open_bank_account'] == "Yes") | (counter['discrimination_credit_prohibited'] == "Yes") :
        return '2'
    if (counter['sign_contract'] == "No") | (counter['register_business'] == "Yes") | (counter['open_bank_account'] == "No") | (counter['discrimination_credit_prohibited'] == "Yes") :
        return '2'
    if (counter['sign_contract'] == "No") | (counter['register_business'] == "Yes") | (counter['open_bank_account'] == "Yes") | (counter['discrimination_credit_prohibited'] == "No") :
        return '2'
    else:
        return 'not 4'

df_entre_periods["law_count"] = df_entre_periods.apply(law_count, axis=1)
```
