```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df_entrepreneur = pd.read_csv('../data/entrepreneurship.csv', engine='python')
df_entrepreneur.shape
```

```python
reportyr_count = df_entrepreneur.reportyr.value_counts()
reportyr_count 
```

```python
def period(counter):
    if (counter['reportyr'] == 1971) | (counter['reportyr'] <= 1975) :
        return '1971-1975'
    if (counter['reportyr'] == 1976) | (counter['reportyr'] <= 1980) :
        return '1976-1980'
    if (counter['reportyr'] == 1981) | (counter['reportyr'] <= 1985) :
        return '1981-1985'
    if (counter['reportyr'] == 1986) | (counter['reportyr'] <= 1990) :
        return '1986-1990'
    if (counter['reportyr'] == 1991) | (counter['reportyr'] <= 1995) :
        return '1991-1995'
    if (counter['reportyr'] == 1996) | (counter['reportyr'] <= 2000) :
        return '1996-2000'
    if (counter['reportyr'] == 2001) | (counter['reportyr'] <= 2005) :
        return '2001-2005'
    if (counter['reportyr'] == 2006) | (counter['reportyr'] <= 2010) :
        return '2006-2010'
    if (counter['reportyr'] == 2011) | (counter['reportyr'] <= 2015) :
        return '2011-2015'
    else:
        return '2016-2021'

df_entrepreneur["period"] = df_entrepreneur.apply(period, axis=1)
```

```python
def period(counter):
    if (counter['reportyr'] == 1971) | (counter['reportyr'] <= 1975) :
        return '1972-1976'
    if (counter['reportyr'] == 1976) | (counter['reportyr'] <= 1980) :
        return '1977-1981'
    if (counter['reportyr'] == 1981) | (counter['reportyr'] <= 1985) :
        return '1982-1986'
    if (counter['reportyr'] == 1986) | (counter['reportyr'] <= 1990) :
        return '1987-1991'
    if (counter['reportyr'] == 1991) | (counter['reportyr'] <= 1995) :
        return '1992-1996'
    if (counter['reportyr'] == 1996) | (counter['reportyr'] <= 2000) :
        return '1997-2001'
    if (counter['reportyr'] == 2001) | (counter['reportyr'] <= 2005) :
        return '2002-2006'
    if (counter['reportyr'] == 2006) | (counter['reportyr'] <= 2010) :
        return '2007-2011'
    if (counter['reportyr'] == 2011) | (counter['reportyr'] <= 2015) :
        return '2012-2016'
    else:
        return '2017-2021'

df_entrepreneur["period"] = df_entrepreneur.apply(period, axis=1)
```
