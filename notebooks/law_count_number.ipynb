```python
def sign_contract_count(counter):
    if (counter['sign_contract'] == 'Yes') | (counter['sign_contract'] == 'Yes') :
        return '1'
    else:
        return '0'

df_law_count["sign_contract_count"] = df_law_count.apply(sign_contract_count, axis=1)
```

```python
def register_business_count(counter):
    if (counter['register_business'] == 'Yes') | (counter['register_business'] == 'Yes') :
        return '1'
    else:
        return '0'

df_law_count["register_business_count"] = df_law_count.apply(register_business_count, axis=1)
```

```python
def open_bank_account_count(counter):
    if (counter['open_bank_account'] == 'Yes') | (counter['open_bank_account'] == 'Yes') :
        return '1'
    else:
        return '0'

df_law_count["open_bank_account_count"] = df_law_count.apply(open_bank_account_count, axis=1)
```

```python
def discrimination_credit_prohibited_count(counter):
    if (counter['discrimination_credit_prohibited'] == 'Yes') | (counter['discrimination_credit_prohibited'] == 'Yes') :
        return '1'
    else:
        return '0'

df_law_count["discrimination_credit_prohibited_count"] = df_law_count.apply(discrimination_credit_prohibited_count, axis=1)
```
